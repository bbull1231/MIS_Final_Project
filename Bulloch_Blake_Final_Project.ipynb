```python
df = df.drop(['color', 'director_name', 'content_rating', 'language', 'country', 'title_year', 'movie_title', 'aspect_ratio', 
              'genres', 'plot_keywords', 'movie_imdb_link', 'actor_1_name', 'actor_3_name', 'actor_2_name', 'num_user_for_reviews', 'cast_total_facebook_likes', 'movie_facebook_likes', 'num_voted_users'], axis=1)
df.head()
```

## building regression 

```python
#define X & y
y = df['imdb_score'] 
X = df[['num_critic_for_reviews','duration','gross','budget','director_facebook_likes','actor_1_facebook_likes']]
```

```python
X.head(2)
```

```python
# build a simple multiple regression
model1 = lm.LinearRegression()
model1.fit(X, y)   
model1_y = model1.predict(X)
```

```python
print 'Coefficients: ', model1.coef_
print "y-intercept ", model1.intercept_
```

```python
coef = ["%.3f" % i for i in model1.coef_]
xcolumns = [ i for i in X.columns ]
zip(xcolumns, coef)
```

```python
print "mean square error: ", mean_squared_error(y, model1_y)
print "variance or r-squared: ", explained_variance_score(y, model1_y)
```

The lower the variance, the lower the score.

```python
def conditions(s):
    if (df['imdb_score'] < 4):
        return 1
    elif (df['imdb_score'] >= 4) and (df['imdb_score'] <= 6):
        return 2
    elif (df['imdb_score'] >= 6) and (df['imdb_score'] <= 8):
        return 3
    else:
        return 4
```

```python
X.var()
```

### K_means

```python
k_means = KMeans(init='k-means++', n_clusters=4, random_state=0)

k_means.fit(X)
```

```python
df_norm = (df - df.mean()) / (df.max() - df.min())
df_norm.head()
```

```python
# clustering analysis using k-means

k_means.fit(df_norm)
```

```python
# find out cluster centers

k_means.cluster_centers_
```

```python
# convert cluster lables to dataframe

df1 = pd.DataFrame(k_means.labels_, columns = ['cluster'])
df1.head()
```
